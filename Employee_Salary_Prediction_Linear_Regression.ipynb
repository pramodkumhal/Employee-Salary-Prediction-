# Feature Engineering

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
Label_Encoder =LabelEncoder()
```

```python
df['Gender_Encode']= Label_Encoder.fit_transform(df['Gender'])
```

```python
df['Degree_Encode']= Label_Encoder.fit_transform(df['Degree'])
```

```python
df['Job_Title_Encode']= Label_Encoder.fit_transform(df['Job_Title'])
```

```python
df['Gender']= Label_Encoder.fit_transform(df['Gender'])
```

```python
df.head()
```

## Feature Scalling

```python
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()
```

```python
df['Age_scaled'] = std_scaler.fit_transform(df[['Age']])
df['Experience_years_scaled'] = std_scaler.fit_transform(df[['Experience_years']])
```

```python
df.head()
```

## Dependent and independent feature

```python
X = df[['Age_scaled','Gender_Encode','Degree_Encode','Job_Title_Encode','Experience_years']]
Y = df['Salary']
```

```python
X.head()
```

## Splitting the data into training and testing

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=42)
```

```python
X_train.head()
```

# Model Development

```python
from sklearn.linear_model import LinearRegression
Linear_regression_model = LinearRegression()
```

## Model training

```python
Linear_regression_model.fit(X_train,Y_train)
```

## Model predictions

```python
y_pred_lr = Linear_regression_model.predict(X_test)
y_pred_lr
```

## Model Evaluation

```python
from sklearn.metrics import accuracy_score, r2_score
from sklearn.metrics import mean_squared_error, mean_absolute_error
```

```python
r2_score(Y_test, y_pred_lr)
```

```python
print(f'Accuracy of the mode = {round(r2_score(Y_test, y_pred_lr),4)*100}%')
```

## Mean absolute Error

```python
round(mean_absolute_error(Y_test,y_pred_lr),2)
print(f'Mean Absolute Error = {round(mean_absolute_error(Y_test, y_pred_lr),2)}')
```

## Mean Squared Error

```python
mse=round(mean_squared_error(Y_test,y_pred_lr),2)
print(f'Mean Absolute Error = {round(mean_squared_error(Y_test, y_pred_lr),2)}')
```

## Root Mean Squared Error

```python
print('Root Mean Squared Error= ',mse**(0.5))
```
